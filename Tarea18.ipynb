```julia
function bifurcacion(x,r,n)                    
    for i in 1:n                              
        x = r*x*(1-x)                          
    end
    return x                                   
    end
```

```julia
function exponenteLyapunov(x,n,r)                       
    l=0    
for i in 1:n                                 
    x = r*x*(1-x)                             
    l = l+log(abs(r*(1-2*x)))                 
    end
    return l/n                             
    end
```

```julia
a=linspace(0,1,1000)
b=linspace(1,4,1000)
list1 =[]
list2 =[]
for i in 1:length(b)
    push!(list1,bifurcacion.(a,b[i],1000))
    push!(list2,ones(length(a))*b[i])
    end
```

```julia
using PyPlot
plot(list2,list1,"o",markersize=0.5)
title("Bifurcacin")
grid()
```

```julia
plot(b,exponenteLyapunov.(0.1,1000,b),"o",markersize=0.5)
title("Exponente de lyaponuv")
grid()
```
